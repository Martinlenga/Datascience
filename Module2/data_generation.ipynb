DATALIFE CYCLE USING SQL

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('students.db')
```

```python
creating_table = """
CREATE TABLE IF NOT EXISTS students_info(
    student_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    phone TEXT, 
    email TEXT UNIQUE NOT NULL,
    course TEXT NOT NULL,
    profile_pic BLOB
)
"""

conn.executescript(creating_table)
```

```python
inserting_data = """
INSERT INTO students_info (name, phone, email, course, profile_pic)
VALUES
('Shaleen', '0790288929', 'shaleen@gmail.com', 'Datascience', 'shaleen.png'),
('Zizo', '0743124541', 'zizogmail.com', 'DevOps', 'zizo.png'),
('Tete', '', 'tete@gmail.com', 'Software Engineering', ''),
('Trey', '0732345234', 'trey@gmail.com', '', 'trey.png');
"""
conn.executescript(inserting_data)
```

```python
pd.read_sql("SELECT * from students_info", conn)
```

```python
transformation = """    
    DELETE FROM students_info
    WHERE course = '';

    UPDATE students_info 
    SET phone = '07009028842', 
        profile_pic = 'update.png'
    WHERE student_id = 3;

    UPDATE students_info
    SET course = 'Artificial Intelligence'
    WHERE student_id = 2;
"""
conn.executescript(transformation)
```
