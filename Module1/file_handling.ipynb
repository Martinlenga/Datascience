```python
# Opening a file
file = open('law.txt', "r")

content = file.read()
print(content)
print("================")

# writing a file (write mode overwrites what was there before)
file = open('law.txt', "w")

file.write("Relationship Between Common Law and Equity.")
file = open('law.txt', "r")
file.read()
```

```python
# appending a file
file = open('law.txt', "a")

file.write("Examples of Equitable Doctrines Applied.")
file = open('law.txt', "r")
file.readlines()
```
