## Operators

### Assignment Operators 

They include:
- addition, subtraction, division, multiplication, exponential, modulus

```python
num1 = 50
num2 = 22

num3 = num1 + num2
num4 = num3 - num1

num5 = num1 / num2
num6 = num1 // num2

num7 = num6 * 4

print(num3)
print(num4)
print(num5)
print(num6)
print(num7)
```

```python
# Comparison Operators (compare 2 values and return whether it is True or False)

current_temp = 28.0
normal_temp = 24.0


print(current_temp > normal_temp)
print(current_temp >= normal_temp)
print(current_temp < normal_temp)
print(current_temp <= normal_temp)
print(current_temp == normal_temp)
print(current_temp != normal_temp)
```

```python
# Logical Operators 
# - AND (all conditions are satisfied) 
# - OR (one condition is satisfied)

# Example: check for admission criteria

age = 17
have_degree =  False

print(age >= 18 and have_degree == True)
print(age >= 18 or have_degree == True)
print(not age >= 18 and not have_degree == True)
```

```python
# Loop (iterates a sequence)

# For loop - done a predetermined number of times
# While loop - repeated until condition is false(can be infinite times)

# list - store multiple items and uses square brackets
# dictionary - store multiple items using curly brackets (key -> value pair)

names = ['Zizo', 'Tete', 'Milito', 'Tesco']

print(names[2])
print(len(names))
print("=============================================")

for name in names:
    print(name)
print("=============================================")

cars ={
    'model': 'BMW',
    'wheels': 5,
    'color': 'red',
    'mileage': 20000,
}

for car in cars:
    print("car details:" + car)

print("=============================================")

students = [{
    'name':'Zizo',
    'age': 19,
    'course': 'Data-science',
    'height': 1.6
}, 

{
    'name':'Trent',
    'age': 17,
    'course': 'DTF',
    'height': 1.7
},

{
    'name':'Kent',
    'age': 19,
    'course': 'Software Dev',
    'height': 1.53
},]

for student in students:
    print(student)
    
print("=============================================")
```

```python
# Dictionaries - stores data in key value pairs

students = {
    'name': 'Martin',
    'age': 22,
    'studentId': '001',
    'gender': 'male',
    'course': 'Data science'
}

print(students['name'])
print(students.get('cohort'))
print("===========================")

# Methods associated with dictionaries
# .keys() .values(), .items()

print(students.keys())
print(students.values())
print(students.items())

print(f"student name: {students['name']}, student age: {students['age']}")
print("===========================")

# . update()
students.update({'name': 'Zizo'})
students.update({'age': 20})
print(students)
print("===========================")

# .pop()
students.pop('studentId')
print(students)

# del
del students['course']
print(students)

# .copy()
students_copy = students.copy()
students_copy.popitem()
print(students_copy)

```
